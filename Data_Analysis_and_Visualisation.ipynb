```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as wn
wn.filterwarnings('ignore')
```

```python
df = pd.read_csv("car_dataset.csv")
```

```python
df.head(10)
```

### Visualization 1: Distribution of Fuel Types

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Fuel Types', data=df, order=df['Fuel Types'].value_counts().index, palette='viridis')
plt.title('Distribution of Fuel Types')
plt.xlabel('Count')
plt.ylabel('Fuel Types')
plt.show()
```

### Visualization 2: Top Companies with the Most Cars

```python
plt.figure(figsize=(12, 6))
top_companies = df['Company Names'].value_counts().head(10)
sns.barplot(x=top_companies.values, y=top_companies.index, palette='coolwarm')
plt.title('Top Companies with Most Cars')
plt.xlabel('Number of Cars')
plt.ylabel('Company Names')
plt.show()
```

### Visualization 3: Relationship between Horsepower and Price

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='HorsePower', y='Price ($)', hue='Fuel Types', data=df, palette='tab10')
plt.title('Horsepower vs Price')
plt.xlabel('Horsepower')
plt.ylabel('Price ($)')
plt.show()
```

### Visualization 4: Average Horsepower by Fuel Types

```python
plt.figure(figsize=(12, 6))
avg_hp_by_fuel = df.groupby('Fuel Types')['HorsePower'].mean().sort_values(ascending=False)
sns.barplot(x=avg_hp_by_fuel.values, y=avg_hp_by_fuel.index, palette='mako')
plt.title('Average Horsepower by Fuel Types')
plt.xlabel('Average Horsepower')
plt.ylabel('Fuel Types')
plt.show()
```

### Visualization 5: Torque vs HorsePower

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Torque', y='HorsePower', hue='Fuel Types', size='Price ($)', data=df, sizes=(20, 200), palette='cubehelix')
plt.title('Torque vs Horsepower')
plt.xlabel('Torque')
plt.ylabel('Horsepower')
plt.show()
```

### Visualization 6: Distribution of Car Prices

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Price ($)'], kde=True, color='blue', bins=20)
plt.title('Distribution of Car Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

### Visualization 7: Car Features vs Price

```python
# Create a 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Car Features vs Price ($)', fontsize=18, weight='bold', y=0.95)

# Chart 1: Battery Capacity vs Price
sns.lineplot(x="Battery Capacity (cc)", y="Price ($)", data=df, ax=axes[0, 0], color="blue", marker="o")
axes[0, 0].set_title("Battery Capacity vs Price", fontsize=14)
axes[0, 0].set_xlabel("Battery Capacity (cc)", fontsize=12)
axes[0, 0].set_ylabel("Price ($)", fontsize=12)

# Chart 2: HorsePower vs Price
sns.lineplot(x="HorsePower", y="Price ($)", data=df, ax=axes[0, 1], color="green", marker="o")
axes[0, 1].set_title("HorsePower vs Price", fontsize=14)
axes[0, 1].set_xlabel("HorsePower", fontsize=12)
axes[0, 1].set_ylabel("Price ($)", fontsize=12)

# Chart 3: Total Speed vs Price
sns.lineplot(x="Total Speed (km/h)", y="Price ($)", data=df, ax=axes[1, 0], color="red", marker="o")
axes[1, 0].set_title("Total Speed vs Price", fontsize=14)
axes[1, 0].set_xlabel("Total Speed (km/h)", fontsize=12)
axes[1, 0].set_ylabel("Price ($)", fontsize=12)

# Chart 4: Seats vs Price
sns.lineplot(x="Seats", y="Price ($)", data=df, ax=axes[1, 1], color="purple", marker="o")
axes[1, 1].set_title("Seats vs Price", fontsize=14)
axes[1, 1].set_xlabel("Seats", fontsize=12)
axes[1, 1].set_ylabel("Price ($)", fontsize=12)

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show the grid
plt.show()
```

# Summary of Analysis and Visualization
The dataset provided contains information about various cars, including their specifications, performance, and prices. Below is the summary of insights derived from the analysis and visualizations:

---
## 1. Fuel Type Distribution
The dataset consists of a variety of fuel types, with Petrol being the dominant one, followed by others like Diesel, Hybrid, and Electric.<br>
Plug-in Hybrid and alternative fuel combinations (e.g., Petrol/Hybrid, Diesel/Petrol) are less common.<br>

## 2. Top Companies with Most Cars
FERRARI has the highest representation in the dataset, showcasing its dominance in luxury and high-performance vehicles.<br>
Other prominent manufacturers include NISSAN, TOYOTA, and LAMBORGHINI, which contribute significantly to the dataset.<br>

## 3. Horsepower vs Price
A strong positive relationship is observed between Horsepower and Price.<br>
Cars with higher horsepower are generally more expensive, indicating their focus on performance and luxury.<br>
Outliers include supercars like Ferrari and Lamborghini, which have extremely high horsepower and price.<br>

## 4. Battery Capacity vs Price
Cars with larger Battery Capacity (cc) tend to be more expensive.<br>
Luxury brands like Rolls Royce and high-performance brands like Ferrari and Lamborghini are associated with higher battery capacities.<br>

## 5. Total Speed vs Price
Cars with higher Total Speed (km/h) are priced higher, reflecting the premium charged for speed-focused engineering.<br>
Supercars dominate this category with speeds over 300 km/h and exceptionally high prices.<br>

## 6. Seats vs Price
Cars with 2 seats (typically sports cars) are significantly more expensive than cars with more seats.<br>
This highlights the premium on performance-oriented vehicles.<br>
Vehicles with 57 seats are more affordable, catering to family or utility-oriented use cases.<br>

## 7. Torque vs Horsepower
A positive correlation exists between Torque and Horsepower.<br>
High-performance vehicles exhibit both high torque and horsepower values.<br>
Luxury cars like Ferrari and Lamborghini dominate the upper range of this relationship.<br>

## 8. Car Prices Distribution
The price distribution is highly skewed, with most cars priced under $500,000.<br>
Supercars like Ferrari, Lamborghini, and Rolls Royce significantly exceed this range, forming the outliers.<br>
