```python
import pandas as pd
import numpy as np
import re, json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('The Ultimate Cars Dataset 2024.csv',encoding='cp1252')
```

```python
def rename_and_clean_speed_column(df):
    """
    Renames the column "Total Speed" to "Total Speed (km/h)"
    and removes " km/h" from its values.

    Parameters:
    df (pd.DataFrame): The input dataframe.

    Returns:
    pd.DataFrame: The updated dataframe with changes.
    """
    if "Total Speed" in df.columns:
        # Rename the column
        df.rename(columns={"Total Speed": "Total Speed (km/h)"}, inplace=True)
        # Remove " km/h" from the values in the column
        df["Total Speed (km/h)"] = df["Total Speed (km/h)"].str.replace(" km/h", "", regex=False)
    else:
        print("Column 'Total Speed' does not exist in the dataframe.")
    
    return df

df = rename_and_clean_speed_column(df)
```

```python
def rename_and_clean_performance_column(df):
    """
    Renames the column "Performance(0 - 100 )KM/H" to "Performance(0 - 100 )KM/H (in sec)"
    and removes " sec" from its values.

    Parameters:
    df (pd.DataFrame): The input dataframe.

    Returns:
    pd.DataFrame: The updated dataframe with changes.
    """
    if "Performance(0 - 100 )KM/H" in df.columns:
        # Rename the column
        df.rename(columns={"Performance(0 - 100 )KM/H": "Performance(0 - 100 )KM/H (in sec)"}, inplace=True)
        # Remove " sec" from the values in the column
        df["Performance(0 - 100 )KM/H (in sec)"] = df["Performance(0 - 100 )KM/H (in sec)"].str.replace(" sec", "", regex=False)
    else:
        print("Column 'Performance(0 - 100 )KM/H' does not exist in the dataframe.")
    
    return df

df = rename_and_clean_performance_column(df)
```

```python
def rename_and_clean_battery_capacity_column(df):
    """
    Renames the column "CC/Battery Capacity" to "Battery Capacity (cc)"
    and removes " cc" from its values.

    Parameters:
    df (pd.DataFrame): The input dataframe.

    Returns:
    pd.DataFrame: The updated dataframe with changes.
    """
    if "CC/Battery Capacity" in df.columns:
        # Rename the column
        df.rename(columns={"CC/Battery Capacity": "Battery Capacity (cc)"}, inplace=True)
        # Remove " cc" from the values in the column
        df["Battery Capacity (cc)"] = df["Battery Capacity (cc)"].str.replace(" cc", "", regex=False).replace(",", "", regex=False)
    else:
        print("Column 'CC/Battery Capacity' does not exist in the dataframe.")
    
    return df

df = rename_and_clean_battery_capacity_column(df)
```

```python
def rename_and_clean_horsepower_column(df):
    """
    Removes " hp" from its values.

    Parameters:
    df (pd.DataFrame): The input dataframe.

    Returns:
    pd.DataFrame: The updated dataframe with changes.
    """
    if "HorsePower" in df.columns:
        # Remove " hp" from the values in the column
        df["HorsePower"] = df["HorsePower"].str.replace(" hp", "", regex=False)
    else:
        print("Column 'HorsePower' does not exist in the dataframe.")
    
    return df

df = rename_and_clean_horsepower_column(df)
```

```python
def rename_and_clean_torque_column(df):
    """
    Removes " Nm" from its values.

    Parameters:
    df (pd.DataFrame): The input dataframe.

    Returns:
    pd.DataFrame: The updated dataframe with changes.
    """
    if "Torque" in df.columns:
        # Remove " hp" from the values in the column
        df["Torque"] = df["Torque"].str.replace(" Nm", "", regex=False)
    else:
        print("Column 'Torque' does not exist in the dataframe.")
    
    return df

df = rename_and_clean_torque_column(df)
```
